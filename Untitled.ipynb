```python
## Data cleaning
```

```python
df_train.head()
```

```python
df_train.text = df_train.text.str.lower()
```

```python
from cleantext import clean
```

```python
df_train['new_text'] = df_train.text.apply(lambda x : clean(x,no_punct=True) ) # the punctuation is removed
```

```python
df_train
```

```python
import nltk
from nltk.corpus import stopwords
stop = set(stopwords.words("english"))
df_train['new_text'] = df_train.new_text.apply(lambda x: " ".join(word for word in x.split() if word.lower() not in stop))
```

```python
from nltk.tokenize import word_tokenize
```

```python
df_train.head(3)
```

```python
df_train['new_text'] = df_train['new_text'].apply(lambda x: word_tokenize(x))
```

```python
df_train['new_text']
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
lemma = WordNetLemmatizer()
```

```python
df_train['lemma'] = df_train['new_text'].apply(lambda x: " ".join(lemma.lemmatize(word) for word in x))
```

```python
df_train
```

```python
df_train.sentiment = df_train.sentiment.map({"neg":0,"pos":1})
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
cv = CountVectorizer()
```

```python
tf = TfidfVectorizer()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_train.iloc[:,3]
y = df_train.iloc[:,1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
cv_train = cv.fit_transform(X_train)
```

```python
cv_test1 = cv.transform(X_test)
```

```python
tf_train = tf.fit_transform(X_train)
```

```python
tf_test = tf.transform(X_test)
```

```python
print("Training data shape:", cv_train.shape)
print("Test data shape:", cv_test1.shape)

# Check vocabulary consistency
print("Vocabulary size after fitting on X_train:", len(cv.vocabulary_))
print("Vocabulary size after transforming X_test:", cv_test1.shape[1])
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(cv_train,y_train)
```

```python
pred = knn.predict(cv_test1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,pred))
```

```python
knn.fit(tf_train,y_train)
```

```python
pred = knn.predict(tf_test)
```

```python
print(accuracy_score(y_test,pred))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

```python
Lr = LogisticRegression()
```

```python
Lr.fit(tf_train,y_train)
```

```python
pred = Lr.predict(tf_test)
```

```python
def metric(y,pred):
    print(accuracy_score(y,pred))
    print(classification_report(y,pred))
    print(confusion_matrix(y_test,pred))
```

```python
metric(y_test,pred)
```

```python
rfc.fit(tf_train,y_train)
```

```python
pred = rfc.predict(tf_test)
```

```python
metric(y_test,pred)
```
